```python
# Chọn kích thước mục tiêu
TARGET_SIZE = (128, 128)  # TODO: Thay đổi theo nhu cầu

# Giải thích: Resize tất cả ảnh về kích thước TARGET_SIZE
def resize_images(images, target_size):
    """
    Resize một mảng ảnh về kích thước mục tiêu
    
    Parameters:
    - images: mảng numpy chứa các ảnh
    - target_size: tuple (height, width) kích thước mục tiêu
    
    Returns:
    - resized_images: mảng ảnh đã được resize
    """
    resized_images = []
    for img in tqdm(images, desc="Resizing images"):
        resized = cv2.resize(img, target_size)
        resized_images.append(resized)
    return np.array(resized_images)

# TODO: Áp dụng resize
# X_train_resized = resize_images(X_train, TARGET_SIZE)
# X_test_resized = resize_images(X_test, TARGET_SIZE)

# print(f"Original shape: {X_train.shape}")
# print(f"Resized shape: {X_train_resized.shape}")
```

### 5.2 Implementation

```python
# Giải thích: Chuyển đổi ảnh RGB sang grayscale
def convert_to_grayscale(images):
    """
    Chuyển đổi ảnh RGB sang grayscale
    
    Parameters:
    - images: mảng numpy chứa ảnh RGB
    
    Returns:
    - grayscale_images: mảng ảnh grayscale
    """
    grayscale_images = []
    for img in tqdm(images, desc="Converting to grayscale"):
        gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
        grayscale_images.append(gray)
    return np.array(grayscale_images)

# TODO: Áp dụng grayscale conversion
# X_train_gray = convert_to_grayscale(X_train_resized)
# X_test_gray = convert_to_grayscale(X_test_resized)

# print(f"Color image shape: {X_train_resized.shape}")
# print(f"Grayscale image shape: {X_train_gray.shape}")
```

## 6. Preprocessing Technique 3: Normalization

### 6.1 Rationale
**Tại sao cần normalization:**
- [Giải thích về tầm quan trọng của normalization trong ML/DL]

**Các phương pháp normalization:**
1. **Min-Max Scaling [0,1]:** [Giải thích]
2. **Scaling to [-1,1]:** [Giải thích]
3. **Standardization (Z-score):** [Giải thích]

### 6.2 Implementation

```python
# Giải thích: Áp dụng các phương pháp normalization khác nhau

# Method 1: Min-Max Scaling to [0,1]
def normalize_minmax_01(images):
    """
    Normalize pixel values to [0,1] range
    """
    return images.astype('float32') / 255.0

# Method 2: Scaling to [-1,1]
def normalize_minmax_11(images):
    """
    Normalize pixel values to [-1,1] range
    """
    return (images.astype('float32') / 127.5) - 1.0

# Method 3: Standardization (Zero mean, unit variance)
def normalize_standardize(images):
    """
    Standardize pixel values (zero mean, unit variance)
    """
    images_flat = images.reshape(-1, images.shape[-1] if len(images.shape) > 3 else 1)
    scaler = StandardScaler()
    images_normalized = scaler.fit_transform(images_flat)
    return images_normalized.reshape(images.shape)

# TODO: Áp dụng các phương pháp normalization
# X_train_norm_01 = normalize_minmax_01(X_train_resized)
# X_train_norm_11 = normalize_minmax_11(X_train_resized)
# X_train_standardized = normalize_standardize(X_train_resized)

# print("Normalization completed!")
# print(f"Min-Max [0,1] - Min: {X_train_norm_01.min():.4f}, Max: {X_train_norm_01.max():.4f}")
# print(f"Min-Max [-1,1] - Min: {X_train_norm_11.min():.4f}, Max: {X_train_norm_11.max():.4f}")
# print(f"Standardized - Mean: {X_train_standardized.mean():.4f}, Std: {X_train_standardized.std():.4f}")
```

## 7. (Optional Bonus) Edge Detection

### 7.1 Rationale
**Tại sao edge detection quan trọng:**
- [Giải thích]

**Các thuật toán edge detection:**
1. **Sobel:** [Giải thích]
2. **Prewitt:** [Giải thích]
3. **Canny:** [Giải thích]

### 7.2 Implementation

```python
# Giải thích: Áp dụng các thuật toán edge detection

def apply_sobel(image):
    """
    Apply Sobel edge detection
    """
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)
    sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)
    sobel = np.sqrt(sobelx**2 + sobely**2)
    return sobel

def apply_canny(image, threshold1=100, threshold2=200):
    """
    Apply Canny edge detection
    """
    if len(image.shape) == 3:
        image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
    
    edges = cv2.Canny(image, threshold1, threshold2)
    return edges

# TODO: Áp dụng edge detection cho một vài ảnh mẫu
# sample_idx = 0
# sample_image = X_train_resized[sample_idx]
# 
# sobel_edges = apply_sobel(sample_image)
# canny_edges = apply_canny(sample_image)
```
